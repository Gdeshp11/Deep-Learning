```python
import torch

from torchvision import transforms

from PIL import Image

img = Image.open("images/red3.jpg")
img
```

```python
# convert image to tensor
convert_tensor = transforms.ToTensor()
img_tensor = convert_tensor(img)
img_tensor[1:2]
```

```python
# get mean of image tensor
mean = torch.mean(img_tensor)
mean
```

```python
# Compute mean of each channel in image
Red_mean, Green_mean ,Blue_mean = torch.mean(img_tensor, dim = [1,2])

# print mean across image channel RGB
print("Red channel mean:", Red_mean)
print("Green channel mean:", Green_mean)
print("Blue channel mean:", Blue_mean)
```
