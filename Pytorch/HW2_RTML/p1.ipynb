```python
from matplotlib import pyplot as plt
import numpy as np
import torch
from torchvision import datasets, transforms
import torch.nn as nn
import pandas as pd
import torch.optim as optim
import matplotlib.pyplot as plt


data_path = "../HW1_RTML/Housing.csv"
data = pd.DataFrame(pd.read_csv(data_path))

data.head() 

```

```python
data_len = len(data) 
num_vars = ['area', 'bedrooms', 'bathrooms', 'stories', 'parking','price'] 
training_data = data[num_vars]
```

```python
training_data_tensor = torch.tensor(training_data.to_numpy())
normalized_data = training_data_tensor / 1000000 #normalization
t_c = normalized_data[:,5] #actual house prices
t_un = normalized_data[:,0:5] #features
```

```python
shuffled_indices = torch.randperm(data_len)
n_val = int (0.2 * data_len)
train_indices = shuffled_indices[:-n_val]
val_indices = shuffled_indices[-n_val:]

train_t_un = t_un[train_indices]
train_t_c = t_c[train_indices]

val_t_un = t_un[val_indices]
val_t_c = t_c[val_indices]
```

```python
def model_1():
    return nn.Sequential(nn.Linear(5, 8),
            nn.Tanh(),
            nn.Linear(8, 1))
```

```python
def training_loop(n_epochs, optimizer, model, loss_fn, t_u_train, t_u_val,t_c_train, t_c_val):
    for epoch in range(1, n_epochs + 1):
        t_p_train = model(t_u_train)
        loss_train = loss_fn(t_p_train, t_c_train)
        t_p_val = model(t_u_val)
        loss_val = loss_fn(t_p_val, t_c_val)
        optimizer.zero_grad()
        loss_train.backward()
        optimizer.step()
        print(f"Epoch {epoch}, Training loss {loss_train.item():.4f},"f" Validation loss {loss_val.item():.4f}")
```

```python
training_model = model_1()
learning_rate = 1e-2
optimizer = optim.SGD(training_model.parameters(), lr=learning_rate)
loss_fn = nn.MSELoss()

training_loop(
n_epochs = 200,
optimizer = optimizer,
model = training_model,
loss_fn = loss_fn,
t_u_train = train_t_un,
t_u_val = val_t_un,
t_c_train = train_t_c,
t_c_val = val_t_c)
```

```python
def model_2():
    return nn.Sequential(nn.Linear(5,8),nn.Tanh(),nn.Linear(8,4),nn.Tanh(),nn.Linear(4,1))
```

```python
training_model = model_2()
learning_rate = 1e-1
optimizer = optim.SGD(training_model.parameters(), lr=learning_rate)
loss_fn = nn.MSELoss()

training_loop(
n_epochs = 200,
optimizer = optimizer,
model = training_model,
loss_fn = loss_fn,
t_u_train = train_t_un,
t_u_val = val_t_un,
t_c_train = train_t_c,
t_c_val = val_t_c)
```
