```python
from torchvision import transforms
preprocess = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(
            mean=[0.485, 0.456, 0.406],
            std=[0.229, 0.224, 0.225]
        )])
```

```python
from PIL import Image
apple = Image.open("./Image-1_apple.jpg") 
parrot = Image.open("./Image-2_parrot.jpg") 
koala =  Image.open("./Image-3_Koala.jpg") 
tree = Image.open("./Image-4_tree.jpg") 
balloon = Image.open("./Image-5_balloon.jpg")
```

```python
img_1 = preprocess(apple)
img_2 = preprocess(parrot)
img_3 = preprocess(koala)
img_4 = preprocess(tree)
img_5 = preprocess(balloon)
```

```python
import torch
batch_1 = torch.unsqueeze(img_1, 0)
batch_2 = torch.unsqueeze(img_2,0)
batch_3 = torch.unsqueeze(img_3,0)
batch_4 = torch.unsqueeze(img_4,0)
batch_5 = torch.unsqueeze(img_5,0)
```

```python
mobilenet_v2.eval()
```

```python
out1 = mobilenet_v2(batch_1)
out2 = mobilenet_v2(batch_2)
out3 = mobilenet_v2(batch_3)
out4 = mobilenet_v2(batch_4)
out5 = mobilenet_v2(batch_5)
```

```python
with open('./imagenet_classes.txt') as f:
    labels = [line.strip() for line in f.readlines()]
```

```python
def analyze_image(arg):
    _, index = torch.max(arg, 1)
    percentage = torch.nn.functional.softmax(arg, dim=1)[0] * 100
    # return labels[index[0]], percentage[index[0]].item()
    _, indices = torch.sort(arg, descending=True)
    return [(labels[idx], percentage[idx].item()) for idx in indices[0][:5]]
```
